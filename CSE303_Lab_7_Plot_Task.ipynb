# Annotate 
Annotate the point xy with text text.

In the simplest form, the text is placed at xy.

Optionally, the text can be displayed in another position xytext. An arrow pointing from the text to the annotated point xy can then be added by defining arrowprops.

\

Learn More:

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.annotate.html

```python
import numpy as np
import matplotlib.pyplot as plt

# Plot
plt.plot([1, 2, 3, 4])

# Label
plt.ylabel('Some Numbers')

# Set x limits
plt.xlim(0,3.0)

# Set y limits
plt.ylim(1.0,4.0)

# Set x ticks
plt.xticks([0,1,2,3],
           ['$0$', '$1$', '$2$','$3$'])

# plt.xticks([0,1,2,3],
#            ['$-\pi$', r'$-\frac{\pi}{2}$', '$0$', '$+\pi/2$'])

# Using mathtext to get Unicode Characters by wrapping the format specifier with $

# Set y ticks
plt.yticks([1,2,3,4],
           ['$1$', '$2$', '$3$','$4$'])


# Dotted line plot
plt.plot([2, 2], [1, 3], color = 'red', linestyle = "--", linewidth = 1.0)

# Annotate
plt.annotate(r'$y = 3$', xy = (2, 3), xytext = (+50, -30), textcoords = 'offset points',
             fontsize = 10, arrowprops = dict(arrowstyle= '->', connectionstyle = "arc3, rad = .5"))

# Title
plt.title("Line")

plt.show()
```

```python
# Code Here
import numpy as np

import matplotlib.pyplot as plot
x = np.arange(-2*np.pi, 2*np.pi,0.01)

plt.xticks([-2*np.pi,-3/2*np.pi,-np.pi,-np.pi/2,0,np.pi/2,np.pi,3/2*np.pi,2*np.pi],['$-2\pi$',r'$-\frac{3\pi}{2}$','$-\pi$',r'$-\frac{\pi}{2}$','$0$',r'$\frac{\pi}{2}$','$\pi$',r'$\frac{3\pi}{2}$','$2\pi$'])

plt.yticks([-1,0,1],
           ['$-1$', '$0$', '$+1$'])

# Set x limits
plt.ylim(-1,1)
amplitude = np.cos(x)

plot.plot(x, amplitude, color='red')

# Dotted line plot
plt.plot([np.pi/3, np.pi/3], [-1, .5], color = 'red', linestyle = "--", linewidth = 1.0)

# Annotate
plt.annotate(r'$cos(\frac{\pi}{3})=\frac{1}{2}$', xy = (np.pi/3, .5), xytext = (+40, -30), textcoords = 'offset points',
             fontsize = 10, arrowprops = dict(arrowstyle= '->', connectionstyle = "arc3, rad = .5"))
plot.title('Cosine wave')

# Display the cosine wave plot

plot.show()
```

# Exercise 2:

Draw a box plot from "adult.data" with age column and grouping by sex column.

Use df.boxplot() for plotting. 

Use Parameters:

Grid should be false.

Also, give appropriate figure size.

Now use Annotate to pinpoint:


*   Median (With Value)
*   Upper Quartile (Q3) (With Value)
*   Lower Quartile (Q1) (With Value)
*   Upper Whisker (With Value)
*   Lower Whisker (With Value)

Also, pinpoint one outlier (If any).

 


```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/drive/MyDrive/Classroom/CSE303/adult.data', header=None)
df.columns = ['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship','race','sex','capital-gain',
              'capital-loss','hours-per-week','native-country', 'salary']
# df.head(10)

# Code Here

ax = df.boxplot(by='sex',column="age", figsize=(10, 8), grid=False);

# Median
median = df.groupby('sex').age.median()
malemedian = median[0]
femalemedian = median[1]


# quartile
Q1 = np.percentile(df.age, 25)
Q2 = np.percentile(df.age, 50)
Q3 = np.percentile(df.age, 75)


#Whisker
upper_whisker = df.groupby('sex').age.max()
lower_whisker = df.groupby('sex').age.min()


# Annotation
#Median
plt.annotate("Median ={}".format(round(malemedian,2)), xy = (1, malemedian), xytext = (+40, -25), textcoords = 'offset points',
             fontsize = 10, arrowprops = dict(arrowstyle= '->', connectionstyle = "arc3, rad = .99"))
plt.annotate("Median ={}".format(round(femalemedian,2)), xy = (2, femalemedian), xytext = (+40, -25), textcoords = 'offset points',
             fontsize = 10, arrowprops = dict(arrowstyle= '->', connectionstyle = "arc3, rad = .99"))

#Q1
plt.annotate("Q1 ={}".format(round(Q1,2)), xy = (1, Q1), xytext = (+40, -30), textcoords = 'offset points',
             fontsize = 10, arrowprops = dict(arrowstyle= '->', connectionstyle = "arc3, rad = .99"))
plt.annotate("Q1 ={}".format(round(Q1,2)), xy = (2, Q1), xytext = (+40, -30), textcoords = 'offset points',
             fontsize = 10, arrowprops = dict(arrowstyle= '->', connectionstyle = "arc3, rad = .99"))

#Q2
plt.annotate("Q2 ={}".format(round(malemedian,2)), xy = (1, malemedian), xytext = (+40, -10), textcoords = 'offset points',
             fontsize = 10, arrowprops = dict(arrowstyle= '->', connectionstyle = "arc3, rad = .99"))
plt.annotate("Q2 ={}".format(round(femalemedian,2)), xy = (2, femalemedian), xytext = (+40, -10), textcoords = 'offset points',
             fontsize = 10, arrowprops = dict(arrowstyle= '->', connectionstyle = "arc3, rad = .99"))

#Q3
plt.annotate("Q3 ={}".format(round(Q3,2)), xy = (1, Q3), xytext = (+40, -30), textcoords = 'offset points',
             fontsize = 10, arrowprops = dict(arrowstyle= '->', connectionstyle = "arc3, rad = .99"))
plt.annotate("Q3 ={}".format(round(Q3,2)), xy = (2, Q3), xytext = (+40, -30), textcoords = 'offset points',
             fontsize = 10, arrowprops = dict(arrowstyle= '->', connectionstyle = "arc3, rad = .99"))

# Upper whisker
plt.annotate("Upper_whisker ={}".format(round(upper_whisker[0],2)), xy = (1, upper_whisker[0]), xytext = (+40, -30), textcoords = 'offset points',
             fontsize = 10, arrowprops = dict(arrowstyle= '->', connectionstyle = "arc3, rad = .99"))
plt.annotate("Lower_whisker ={}".format(round(upper_whisker[1],2)), xy = (2, upper_whisker[1]), xytext = (+40, -30), textcoords = 'offset points',
             fontsize = 10, arrowprops = dict(arrowstyle= '->', connectionstyle = "arc3, rad = .99"))


#Lower whisker
plt.annotate("Upper_whisker ={}".format(round(lower_whisker[0],2)), xy = (1, lower_whisker[0]), xytext = (+40, -30), textcoords = 'offset points',
             fontsize = 10, arrowprops = dict(arrowstyle= '->', connectionstyle = "arc3, rad = .99"))
plt.annotate("Lower_whisker ={}".format(round(lower_whisker[1],2)), xy = (2, lower_whisker[1]), xytext = (+40, -30), textcoords = 'offset points',
             fontsize = 10, arrowprops = dict(arrowstyle= '->', connectionstyle = "arc3, rad = .99"))
```

```python
# Code Here
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))

# Subplots
ax1 = df.plot(x='age', y='education-num',color=['blue'],legend=None, ax=axes[0,0])
ax2 = df.plot(x='age', y='capital-gain', color=['blue'],legend=None, ax=axes[0,1])
ax3 = df.plot(x='age', y='capital-loss', color=['green'],legend=None, ax=axes[1,0])
ax4 = df.plot(x='age', y='hours-per-week',color=['black'],legend=None, ax=axes[1,1])


#Title
ax1.title.set_text('Age and Education_num')
ax1.set(xlabel=None)

ax2.title.set_text('Age and Capital gain')
ax2.set(xlabel=None)

ax3.title.set_text('Age and Capital Loss')
ax3.set(xlabel=None)

ax4.title.set_text('Age and Hours-Per-Week')
ax4.set(xlabel=None)
plt.show()
```
