# Problem 1:

Plot Sine and Cosine Graph. 

Then show Cos(0) and Sin(150) degree value using Annotate. 

The plot graph should look like this.

image.png

```python
# Code Here
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-2*np.pi, 2*np.pi, 0.01)
y = np.arange(-2*np.pi, 2*np.pi, 0.01)
plt.xticks([-2*np.pi, -3/2*np.pi, -np.pi, -np.pi/2, 0, np.pi/2, np.pi, 3/2*np.pi, 2*np.pi],['$-2\pi$', r'$-\frac{3\pi}{2}$','$-\pi$', r'$-\frac{\pi}{2}$', '$0$', r'$\frac{3\pi}{2}$','$\pi$',r'$\frac{3\pi}{2}$','$2\pi$'])
plt.yticks([-1,0,1],['$-1$', '$0$', '$+1$'])

#set xlimits
plt.ylim(-1,1)
amplitude = np.cos(x)
amplitude_y= np.sin(y)

plt.plot(x, amplitude, color='red')
plt.plot(y, amplitude_y, color='blue')

#Dotted line plot
plt.plot([0, 0], [-1, 1], color='blue', linestyle="--", linewidth=1.0)
plt.plot([2.6, 2.6], [ -1, .5], color = 'blue', linestyle ="--", linewidth=1.0)
plt.title('Sine and Cosine wave')

plt.annotate(r'$cos(0)=1$', xy =(0, 1),xytext=(-50,-60), textcoords = 'offset points',
             fontsize=10, arrowprops = dict(arrowstyle = "->", connectionstyle="arc3, rad = .5"))

plt.annotate(r'$sin(\frac{5\pi}{6})=\frac{1}{2}$', xy = (2.6, 1/2), xytext=(+100,30), textcoords ='offset points',
             fontsize = 10, arrowprops = dict(arrowstyle = '->', connectionstyle="arc3 , rad = .5"))
plt.legend(['cosine', 'sine'])
plt.show()
```
