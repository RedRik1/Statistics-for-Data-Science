```python
# Code Here
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('/content/drive/MyDrive/Classroom/CSE303/company_sales_data.csv')
months = df['month_number'].tolist()
rsum = df.sum(axis=1)
ax = rsum.plot(linestyle='--',color='red',marker = 'o',
         markerfacecolor = 'black', markersize = 8)

ax.set_xlabel('Month number')
ax.set_ylabel('Profit in Dollar')
ax.set_title('Company sales data of last year')
ax.legend(['Profit data of last year'],loc='lower right')
plt.xticks(months)
plt.show()
```

# Exercise 5:
Plot Sine and Cosine Graph.

The plot graph should look like this.

image.png

```python
# Code Here
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(42)
x = np.arange(-np.pi, np.pi,0.1)
y = np.arange(-np.pi, np.pi,0.1)
sin_x = np.sin(x)
cos_x = np.cos(y)

# Plot a sine wave using time and amplitude obtained for the sine wave

plt.plot(x, sin_x, color='blue')
plt.plot(y, cos_x, color='green')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend(['cosine', 'sine'], loc='upper left')
plt.show()
```
